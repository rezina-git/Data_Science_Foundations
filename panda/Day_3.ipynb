```python
import pandas as pd
```

```python
df_employees=pd.DataFrame({
'emp_id':[1,2,3,4],
'name':['Reju','Ram','Hari','Shital'],
'dep_id':[101,102,103,104]
})
df_employees
```

```python
df_departments=pd.DataFrame({
    'dep_id':[101,102,105],
    'dep_name':['HR','Finance','IT']
})
df_departments
```

## Inner Join

```python
pd.merge(df_employees,df_departments,on='dep_id',how='inner') #gives details of Reju and ram.i,e. common dep_id.
```

### Left Join

```python
pd.merge(df_employees,df_departments,on='dep_id',how='left') #gives details of 1st tebale all along with dep name of Reju and ram as well.
```

### Right Join

```python
pd.merge(df_employees,df_departments,on='dep_id',how='right') #gives details of 2nd table all along with name of Reju and ram as well.
```

### Full Outer Join

```python
pd.merge(df_employees,df_departments,on='dep_id',how='outer') #gives details of all like union
```

### Different columns name

```python
pd.merge(df_employees, df_departments, how='inner', left_on='dep_id', right_on='dep_id')
```

# loc and iloc
In Pandas, both iloc and loc are used to select rows and columns from a DataFrame, but they work differently.

✅ iloc (Index Location):
Uses numbers (integer positions) to select rows/columns.

Think of it like saying: “Give me the first, second, third row.”

✅ loc (Label Location):
Uses labels (names/index values) to select rows/columns.

Think of it like saying: “Give me the row with index name 'A'.”

```python
df=pd.DataFrame({
    'Name':['Alice','Bob','Charlie'],
    'Age':[25,30,39]
},index=['a','b','c'])
df
```

### Loc

```python
df.loc['a']
```

### iloc

```python
df.iloc[2,1]
```
