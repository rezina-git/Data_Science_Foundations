### Student performance visualization using matplotlib

##### Calculate the total marks of each students and plot them in piechart.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
students = ['Binita','Sunita','Maya','Kritika','Rejina']
subjects = ['English','Math','science','Social','Nepali']
Marks = [
    [30,35,40,80,90],
    [60,50,55,70,85],
    [45,40,60,75,80],
    [70,75,80,85,90],
    [65, 70, 75, 80, 85]
]
```

```python
#Calculate total marks for each students
total_marks = [sum(m) for m in Marks]
total_marks
```

```python
plt.figure(figsize=(5,3))
plt.bar(students,total_marks,color='green')
plt.title('Total Marks of Students')
plt.xlabel('students')
plt.ylabel('Total marks')
plt.grid( linestyle='--', alpha= 0.8)
plt.show()
```

### Determine how many students passed or faioled based on subject_wise pass criteria

```python
# pass fail ratio


# Initialize pass_count to 0 so we can count how many students passed
# (We must do this before using pass_count += 1)
pass_count = 0

# Loop through each student's marks in the Marks list
# (We want to check if each student passed or failed)
for m in Marks:
    
    # Check if the student has scored at least 35 in all subjects
    # (If any subject is below 35, the student is considered failed)
    if all(score >= 35 for score in m):
        
        # If the student passed all subjects, increase the pass count by 1
        # (This helps us keep track of total passed students)
        pass_count += 1

# Calculate the number of failed students
# (Total students - passed students gives us the number of failed)
fail_count = len(students) - pass_count

# Create a new figure for the pie chart with specified size
# (This sets the size of the output chart for better visibility)
plt.figure(figsize=(5,4))

# Create a pie chart showing pass vs fail counts
# (We visualize the result to quickly understand the comparison)
# labels=['pass','fail'] shows labels on the chart slices
# autopct='%1.2f%%' shows the percentage values on each slice
plt.pie([pass_count, fail_count], labels=['pass', 'fail'], autopct='%1.2f%%')

#             This is formatting:

# %1.2f → show number with 2 decimal places

# %% → add % symbol

# So 42.8571428571 becomes → 42.86%


# Add a title to the pie chart
# (Gives context to what the chart is showing)
plt.title('pass vs fail')

# Display the chart
# (This line is necessary to actually show the chart output)
plt.show()
```

## Temperature trends of Nepal's cities

#### lets visualize the monthly average temperatures of three major cities in Nepal — Kathmandu, Pokhara, and Ilam using a line graph.

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
kathmandu = [10, 12, 16, 20, 23, 25, 24, 24, 22, 19, 15, 11]
pokhara =   [13, 15, 18, 22, 25, 27, 26, 26, 24, 21, 17, 14]
ilam =      [11, 13, 17, 21, 24, 26, 25, 25, 23, 20, 16, 12]
plt.figure(figsize=(6,5))
plt.plot(months,kathmandu,label='kathmandu', marker='>',color='green')
plt.plot(months, pokhara, label="Pokhara", marker='s',color='Red')
plt.plot(months, ilam, label="Ilam", marker='^')

plt.title("Average Monthly Temperature in Nepal Cities")
plt.xlabel("Months")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
brands = ['Samsung', 'Xiaomi', 'iPhone', 'Realme']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']

samsung_sales = [150, 180, 200, 170, 190, 210]
xiaomi_sales = [140, 160, 180, 160, 170, 190]
iphone_sales = [130, 150, 170, 140, 160, 180]
realme_sales = [120, 140, 160, 130, 150, 170]

n_months = len(months)
n_brands = len(brands)
x = np.arange(n_months)

width = 0.2
plt.figure(figsize=(10,4))
plt.bar(x - 1.5*width, samsung_sales, width, label='Samsung', color='dodgerblue')
plt.bar(x - 0.5*width, xiaomi_sales, width, label='Xiaomi', color='orange')
plt.bar(x + 0.5*width, iphone_sales, width, label='iPhone', color='green')
plt.bar(x + 1.5*width, realme_sales, width, label='Realme', color='purple')

plt.title('Monthly Mobile Brand Sales Comparison (Jan - Jun)')
plt.xlabel('Months')
plt.ylabel('Sales Units')
plt.xticks(x, months)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
