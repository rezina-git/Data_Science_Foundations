## Line plot : line plot are effective for showing relationships,patterns and trends between two numerical variables

# axline and axvline
#### axvline  Axes Vertical Line
#### axhline  Axes Horizontal Line

 ###  Purpose of axvline and axhline
        -axvline(x) draws a vertical line at x-axis value = x
        -axhline(y) draws a horizontal line at y-axis value = y
 Why use them?
    -To highlight specific values.
    -Show cutoff points or important thresholds.
    -Make your plot more informative.

```python
plt.figure(figsize=(10, 4))  # Set figure size

plt.xticks(df['index'])  # Set custom x-axis ticks using 'index' column

# Plot Rohit Sharma's data
plt.plot(df['index'], df['RG Sharma'],
         marker='>', markersize=10, color='green',
         linewidth=4, linestyle='--', label='rohit')

# Plot Virat Kohli's data
plt.plot(df['index'], df['V Kohli'],
         marker='*', markersize=10, color='blue',
         linewidth=3, linestyle='-', label='kohli')

plt.legend(loc='lower center')   # Place legend at bottom center
plt.title("rohit sharma")        # Set plot title
plt.xlabel("season")             # Label for x-axis
plt.ylabel("runs")               # Label for y-axis

# Highlight important points
plt.axvline(2010, color='yellow')  # Vertical line at season 2010
plt.axhline(850, color='yellow')   # Horizontal line at runs = 850

plt.grid()        # Add grid for better readability
plt.show()        # Display the plot
```
